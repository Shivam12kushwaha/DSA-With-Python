#### Linked List Implementation

```python
# Constructor to initialize a new node with data
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
    
```

#### Printing the linked list

```python
temp = head
while temp is not None:
    print(temp.data, end = ' ' )
    temp = temp.next
```

#### Singly Linkedlist

```python
class SinglyLinkedList:
    def __init__(self, head=None):
        self.head = head


    # Insert at the end
    def InsertAtEnd(self, value):
        temp = Node(value)
        
        if self.head is not None:
            t1 = self.head
            while (t1.next is not None):
                t1 = t1.next
            t1.next = temp
        else:
            self.head = temp


    # Insert at the beginning
    def InsertAtBegin(self, value):
        temp = Node(value)
        temp.next = self.head
        self.head = temp

    # Insert in  between 

    def InsertAtMid(self, value, x):
        temp = Node(value)
        t1 = self.head

        while t1.next is not None:
            if (t1.data  == x):
                temp.next = t1.next
                t1.next = temp
            t1 = t1.next 


    # Delete a node
    def deleteNode(self, value):
        t1 = self.head
        # Corner Case: Deleting first node
        prev = t1
        if (t1.data == value):
            self.head = t1.next

        # Middle Case
        while (t1.next is not None):
            if (t1.data) == value:
                prev.next = t1.next
                break
            else:
                prev = t1
                t1 = t1.next
        
        # Corner Case: Deleting last node
        if (t1.data == value):
            prev.next = None

    # Printing Linked List
    def printLL(self):
        t1 = self.head

        while t1.next is not None:
            print(t1.data)
            t1 = t1.next
        print(t1.data)
```

#### Exmaples

```python
obj = SinglyLinkedList()
obj.InsertAtEnd(50)
obj.InsertAtEnd(40)
obj.InsertAtEnd(30)
obj.InsertAtBegin(5)
obj.InsertAtMid(10,40)
obj.printLL()
```
